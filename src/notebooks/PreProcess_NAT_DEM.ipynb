# Pre-Procesamiento de texto

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import spacy 
import random


random.seed(0)
```

```python
from hydra import initialize, compose

with initialize(version_base=None, config_path="../conf"):
    cfg=compose(config_name="main.yaml")
```

```python
# Data import
# Relative Path of the dataset, change for your dataset

data_naturaleza = pd.read_csv('../../data/raw_dataset/normalize/Naturaleza_normalize.csv', delimiter=",")

data_democracia = pd.read_csv('../../data/raw_dataset/normalize/Democracia_Normalize.csv', delimiter=",")

nlp = spacy.load('es_core_news_sm')
```

# Naturaleza

```python
concept = data_naturaleza.iloc[:,0].value_counts()

names = list(concept.keys())
values = list(concept)

plt.barh(names, values)
```

```python
cod = data_naturaleza.iloc[:,2].value_counts()
reduce_cod = cod[cod>25]

names = list(reduce_cod.keys())
values = list(reduce_cod)

plt.barh(names, values)
```

```python
descriptions = data_naturaleza.iloc[:,1].copy()

number_words = np.zeros(len(descriptions))

i=0
for d in descriptions:
    number_words[i]= len(nlp(d))
    i+=1


print("min:", number_words.min())
print("max:", number_words.max())
print("mean:", number_words.mean())
print("std:", number_words.std())
print("\n\nTotal de palabras", number_words.sum())
```

```python
cant_codes = data_naturaleza.iloc[:,2].unique().shape[0]
data_len = data_naturaleza.shape[0]

print(f"la cantidad de codigos posibles es: {cant_codes:,}", )
print(f"la cantidad de datos es: {data_len:,}", )

print(f"Proporcin de conceptos por cantidad de datos: {data_len/cant_codes}")
```
