```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotnine import *
from pathlib import Path

```

```python
d_name = "Democracia"
data=pd.read_csv(f"../../../data/resultados_ep1_{d_name}.csv")
data["funca"]=data["model"]+"\n"+data["type"]
data["Min"] = data["mean"]-data["std"]
data["max"] = data["mean"]+data["std"]
data
```

```python
temp=(ggplot(data)+aes(x="funca",y="mean",color="E-type",shape="E-type",ymin="Min",ymax="max")+theme_bw()
+theme(legend_position=(0.925,0.2))
+geom_point(size=4)
+labs(title='', x="", y="Accuracy")
+theme(axis_text_y = element_text(size=15))+theme(axis_title_y = element_text(size=20))
+theme(axis_text_x = element_text(size=15))
+theme(title = element_text(size=25))
+theme(legend_title = element_text(size=20))+theme(legend_background = element_rect(color="black",size=0.5))
+theme(legend_text = element_text(size=15))
+geom_linerange(size=1)
+scale_color_manual(name = "Accuracy",values=["#D83232","#3837D9"])
+scale_shape_manual(name = "Accuracy",values = ["s",'o'])
)
temp.save(filename=f'../../../data/results/IMG/{d_name}_resultados_exp1.png',width=15, height=7)
temp
```
