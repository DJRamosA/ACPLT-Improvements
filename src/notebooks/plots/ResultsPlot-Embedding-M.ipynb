```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotnine import *
from pathlib import Path
# import sys
# !{sys.executable} -m pip install plotnine
```

```python
data=pd.read_csv("funca.csv")
data
data["funca"]=data["model"]+"\n"+data["type"]
data
```

```python
temp=(ggplot(data)+aes(x="funca",y="mean",color="E-type",shape="E-type",ymin="Min",ymax="max")+theme_bw()+geom_point(size=4)
+theme(legend_position=[0.925,0.2])+labs(title="", x="", y="Accuracy")
+theme(axis_text_y = element_text(size=15))+theme(axis_title_y = element_text(size=20))
+theme(axis_text_x = element_text(size=15))
+theme(title = element_text(size=25))
+theme(legend_title = element_text(size=20))+theme(legend_background = element_rect(color="black",size=0.5))
+theme(legend_text = element_text(size=15))
+geom_linerange(size=1)
+scale_color_manual(name = "Accuracy",values=["#D83232","#3837D9"])
+scale_shape_manual(name = "Accuracy",values = ["s",'o'])
)
temp.save(filename='funca.png',width=15, height=7)
#ggsave(filename='funca.png',data=temp) #, device, width, height, units, dpi = 300, limitsize = TRUE)
```
