```python
import pandas as pd
import numpy as np
import time
import glob
import re
```

```python
def correction_csv(file:str):
    extract_property = r'PROP: (\w+\W*\w*),'
    extract_codification= r'CODE: (\w+\W*\w*)\s'
    df = []
    with open(file, 'r', newline='') as f:
        line = f.readline()
        need_property = True
        prop_code_pair = ['','']
       
        while line:
            if need_property:
                tmp = re.match(extract_property, line)
                if tmp:
                    # print(tmp.groups()[0])
                    prop_code_pair[0] = tmp.groups()[0]
                    need_property = False
            else:
                tmp = re.match(extract_codification, line)
                # print(line, "end")
                if tmp: 
                    # print(tmp.groups()[0])
                    prop_code_pair[1] = tmp.groups()[0]
                    need_property = True
                    df.append(prop_code_pair.copy())
            line = f.readline()

        f.close()
    return pd.DataFrame(df, columns=('property', 'code'))


def find_pattern_prop(p):
    property_identifier = r"\'(\w.*\w)\'"
    return re.match(property_identifier, p).group(1)

def find_pattern_cod(p):
    property_identifier = r" \'(\w.*\w)\'"
    return re.match(property_identifier, p).group(1)
```
